```python
from pyspark.sql.functions import current_timestamp, input_file_name
```

```python
# 1. Caminhos
caminho_landing_ibge = "/Volumes/oftalmo_sus/00_landing/raw_files/IBGE/municipios_ibge.csv"
nome_tabela_bronze = "oftalmo_sus.01_bronze.bronze_dim_municipios_ibge"
```

```python
# 2. Leitura com inferência de schema e cabeçalho
df_raw_ibge = spark.read.csv(caminho_landing_ibge, header=True, inferSchema=True)
display(df_raw_ibge.limit(10))
```

```python
# 3. Adição de metadados padrão Bronze
from pyspark.sql.functions import col

df_bronze_ibge = (
    df_raw_ibge
    .withColumn("data_ingestao_bronze", current_timestamp())
    .withColumn("arquivo_origem", col("_metadata.file_path"))
)
```

```python
# 4. Carga Delta
(
    df_bronze_ibge.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(nome_tabela_bronze)
)

print("✅ Dimensão Geográfica salva na camada Bronze!")
```
