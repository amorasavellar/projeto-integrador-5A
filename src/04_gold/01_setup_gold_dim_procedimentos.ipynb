```python
%skip
%sql
CREATE OR REFRESH MATERIALIZED VIEW oftalmo_sus.03_gold.mv_dim_procedimentos
COMMENT 'Dimensão dos procedimentos oftalmo SIGTAP'
AS
SELECT *
FROM oftalmo_sus.02_silver.tb_dim_sigtap_oftalmo;
```

```python
# Lendo a tabela Silver
df_silver_sigtap = spark.read.format("delta").load("oftalmo_sus.02_silver.tb_dim_sigtap_oftalmo")

# Gravando fisicamente na Gold no banco de dados 'default' do Databricks Free
df_silver_sigtap.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("oftalmo_sus.03_gold.dim_procedimentos")

print("Dimensão de Procedimentos gravada com sucesso!")
```
