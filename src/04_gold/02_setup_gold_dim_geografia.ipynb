```python
%skip
%sql
CREATE OR REFRESH MATERIALIZED VIEW oftalmo_sus.03_gold.mv_dim_geografia
COMMENT 'Dimensão geográfica dos municípios IBGE'
AS 
SELECT 
    codigo_ibge_6, 
    nome_municipio,
    uf,
    latitude,
    longitude
FROM oftalmo_sus.02_silver.tb_dim_municipios_ibge;
```

```python
# Lendo a Silver do IBGE
df_silver_ibge = spark.read.format("delta").load("oftalmo_sus.02_silver.tb_dim_municipios_ibge")

# Selecionando apenas as colunas da modelagem Star Schema
df_gold_geografia = df_silver_ibge.select(
    "codigo_ibge_6", 
    "nome_municipio",
    "uf",
    "latitude",
    "longitude"
)

# Gravando fisicamente na Gold
df_gold_geografia.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("oftalmo_sus.03_gold.dim_geografia")

print("Dimensão de Geografia gravada com sucesso!")
```
