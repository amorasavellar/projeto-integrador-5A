```python
import urllib.request
import os
```

```python
# 1. Configurando o Caminho da Landing
caminho_landing_ibge = "/Volumes/oftalmo_sus/00_landing/raw_files/IBGE"
os.makedirs(caminho_landing_ibge, exist_ok=True)
```

```python
# 2. A Fonte de Dados (Repositório Open Source consolidado de Lat/Long do IBGE)
url_municipios = "https://raw.githubusercontent.com/kelvins/Municipios-Brasileiros/main/csv/municipios.csv"
caminho_destino_csv = f"{caminho_landing_ibge}/municipios_ibge.csv"
```

```python
# 3. Fazendo o Download
print("📥 Conectando e baixando a base geográfica de municípios...")
try:
    urllib.request.urlretrieve(url_municipios, caminho_destino_csv)
    print("✅ Download concluído! Coordenadas aterrissadas com sucesso.")
except Exception as e:
    print(f"❌ Erro ao baixar o arquivo CSV: {e}")
```

```python
# 4. Validação
arquivos_finais = os.listdir(caminho_landing_ibge)
print(f"📂 Arquivos atualmente na Landing IBGE: {arquivos_finais}")
```
