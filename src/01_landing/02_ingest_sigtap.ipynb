```python
from ftplib import FTP
import zipfile
import os
```

```python
volume = "/Volumes/oftalmo_sus/00_landing/raw_files"
path = f"{volume}/SIGTAP"
os.makedirs(path, exist_ok=True)

path
```

```python
# 2. Conectando e Navegando no FTP
print("🔌 Conectando ao FTP do DATASUS...")
ftp = FTP('ftp2.datasus.gov.br')
ftp.login() # O DATASUS permite login anônimo
ftp.cwd('/pub/sistemas/tup/downloads/')
```

```python
print("🔎 Buscando a tabela mais recente disponível no servidor...")
arquivos_ftp = ftp.nlst() 
tabelas_zip = [f for f in arquivos_ftp if f.startswith('TabelaUnificada') and f.endswith('.zip')]
```

```python
tabelas_zip.sort(reverse=True)
arquivo_mais_recente = tabelas_zip[0]
print(f"🎯 Arquivo alvo encontrado: {arquivo_mais_recente}")
```

```python
caminho_destino_zip = f"{path}/{arquivo_mais_recente}"
caminho_destino_zip
```

```python
print(f"📥 Baixando {arquivo_mais_recente} (tem cerca de 20MB)")
with open(caminho_destino_zip, 'wb') as arquivo_local:
    ftp.retrbinary(f'RETR {arquivo_mais_recente}', arquivo_local.write)

ftp.quit() 
print("✅ Download concluído com sucesso!")
```

```python
print("📦 Extraindo tb_procedimento.txt")
try:
    with zipfile.ZipFile(caminho_destino_zip, 'r') as zip_ref:
        zip_ref.extract("tb_procedimento.txt", path=path)
        print("✅ Arquivo 'tb_procedimento.txt' extraído com sucesso na Landing!")
except Exception as e:
    print(f"❌ Erro ao extrair o ZIP: {e}")

print(f"📂 Arquivos na Landing SIGTAP: {os.listdir(path)}")
```
